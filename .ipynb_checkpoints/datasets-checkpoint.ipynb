```python
# import
import pandas as pd
```

```python
# read crime data
crime_df_1 = pd.read_excel('data/crime/Crimes_Against_Persons_Offenses_Offense_Category_by_State_2020.xls.xlsx')
crime_df_2 = pd.read_excel('data/crime/Crimes_Against_Property_Offenses_Offense_Category_by_State_2020.xls.xlsx')
crime_df_3 = pd.read_excel('data/crime/Crimes_Against_Society_Offenses_Offense_Category_by_State_2020.xls.xlsx')

# read business ranking data
business_df_future = pd.read_csv('data/sales/Future50.csv')
business_df_top100 = pd.read_csv('data/sales/Independence100.csv')
business_df_top250 = pd.read_csv('data/sales/Top250.csv')

# read cpi data
```

## crime data
- crime_df_1: crime against persons offenses
    - in this dataset, different kinds of offenses related to persons are recorded among 46 states. 
    - crime such as homicide, human trafficking, etc.
- crime_df_2: crime against property offenses
    - in this dataset, different kinds of offenses related to property are recorded among 46 states. 
    - crime such as fraud, arson, etc.
- crime_df_3: crime against society offenses
    - in this dataset, different kinds of offenses related to society are recorded among 46 states. 
    - crime such as prostitution, drug, etc.
    
- source: https://www.fbi.gov/how-we-can-help-you/more-fbi-services-and-information/ucr
- as shown below, some of the useful columns to be joined will be "State". By joining this column with other datasets, we can then analyze the correlation between crimes that happened in particular states and sales/business ranking in the same state as well. More, we could even analyze whether sales/ranking are affected by certain types of crime. 

```python
# persons crime
crime_df_1 = crime_df_1.dropna()
columns_1 = ['State', 'Number of Participating Agencies', 'Population Covered', \
             'Total Offenses', 'Assault Offenses' , 'Homicide Offenses', \
             'Human Trafficking', 'Kidnapping/Abduction', 'Sex Offenses']

crime_df_1.set_axis(columns_1, axis=1, inplace=True)
crime_df_1.head(10)
```

```python
# property crime
crime_df_2 = crime_df_2.dropna()
columns_2 = ['State', 'Number of Participating Agencies', 'Population Covered', \
             'Total Offenses', 'Arson' , 'Bribery', 'Burglary/Breaking & Entering', \
             'Counterfeiting/Forgery', 'Destruction/Damage/Vandalism', 'Embezzlement', \
            'Extortion/Blackmail', 'Fraud Offenses', 'Larceny/Theft Offenses', 'Motor Vehicle Theft', \
            'Robbery', 'Stolen Property Offenses']

crime_df_2.set_axis(columns_2, axis=1, inplace=True)
crime_df_2.head(10)
```

```python
# society crime
crime_df_3 = crime_df_3.dropna()
columns_3 = ['State', 'Number of Participating Agencies', 'Population Covered', \
             'Total Offenses', 'Animal Cruelty', 'Drug/Narcotic Offenses', 'Gambling Offenses', \
            'Pornography/Obscene Material', 'Prostitution Offenses', 'Weapon Law Violations']

crime_df_3.set_axis(columns_3, axis=1, inplace=True)
crime_df_3.head(10)
```

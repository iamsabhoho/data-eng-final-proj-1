```python
# import
import pandas as pd
```

```python
# read crime data
crime_df_3 = pd.read_excel('/Users/sabrinaho/Documents/GitHub/data-eng-final-proj-1/data/crime/Crimes_Against_Society_Offenses_Offense_Category_by_State_2020.xls.xlsx')

# read business ranking data
business_df_top250 = pd.read_csv('/Users/sabrinaho/Documents/GitHub/data-eng-final-proj-1/data/sales/Top250.csv')
```

## crime 3 clean

```python
# society crime
crime_df_3 = crime_df_3.dropna()
columns_3 = ['State', 'Number of Participating Agencies', 'Population Covered', \
             'Total Offenses', 'Animal Cruelty', 'Drug/Narcotic Offenses', 'Gambling Offenses', \
            'Pornography/Obscene Material', 'Prostitution Offenses', 'Weapon Law Violations']

crime_df_3.set_axis(columns_3, axis=1, inplace=True)
crime_df_3.head(10)
```

## business top250 clean

```python
business_df_top250.head()
```

```python
wiki = pd.read_html('https://en.wikipedia.org/wiki/List_of_restaurant_chains_in_the_United_States')

for restaurant in business_df_top250.Restaurant:
    for table in wiki:
        try:
            hq = table[table.Name.str.lower() == restaurant.lower()].Headquarters.values[0]
            if len(hq) > 0:
                ind = business_df_top250.index[business_df_top250.Restaurant == restaurant].tolist()[0]
                business_df_top250.at[ind, 'Headquarters'] = hq
        except:
            continue
```

```python
business_df_top250.head()
```

```python
business_df_top250.drop('Content', axis=1, inplace=True)
```

```python
business_df_top250.dropna(inplace=True)
```

```python
lst = business_df_top250.Headquarters.str.split(', ')
city = []
state = []

for i in lst:
    city.append(i[0])
    state.append(i[1])
    
business_df_top250['City'] = city
business_df_top250['State'] = state
```

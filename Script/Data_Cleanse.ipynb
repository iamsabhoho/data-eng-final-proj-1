## Data cleansing

```python
#Script removes duplicates and NA values

def dataCleanse(df):
    df = df.drop_duplicates()
    df = df.dropna()
    
dataCleanse(future50_df)
dataCleanse(independence100_df)
```

```python
#Separate Location to City and State columns

future50_df[['City', 'State']] = future50_df['Location'].str.split(', ', expand=True)
future50_df['City'] = future50_df['City'].str.strip()
future50_df['State'] = future50_df['State'].str.strip()
independence100_df['City'] = independence100_df['City'].str.strip()
independence100_df['State'] = independence100_df['State'].str.strip()

display(future50_df.head(5))
```

```python
future50_df = future50_df.drop('Location',axis=1)
```

```python
#Get unique States
state_df = pd.concat([future50_df['State'],independence100_df['State']])
print(sorted(state_df.unique()))
```

```python
state_dict = {'Ariz.':'Arizona', 'Ark.':'Arkansas', 'Calif.':'California', 'Colo.':'Colorado', 'D.C.':'D.C.', 'Fla.':'Florida',
              'Ga.':'Georgia', 'Ill.':'Illinois', 'Ind.':'Indiana', 'Ky.':"Kentucky", 'Mass.':'Massachusetts', 'Mich.':'Michigan', 
              'Mo.':'Missouri', 'N.C.':'North Carolina', 'N.J.':'New Jersey', 'N.Y.':'New York', 'Neb.':'Nebraska',
              'Nev.':'Nevada','Ohio':'Ohio', 'Ore.':'Oregon', 'Pa.':'Pennsylvania', 'S.C.':'South Carolina',
              'Tenn.':'Tennessee', 'Texas':'Texas', 'Va.':'Virginia', 'Wash.':'Washington'}

future50_df['State'] = future50_df['State'].map(state_dict)
independence100_df['State'] = independence100_df['State'].map(state_dict)
              
```

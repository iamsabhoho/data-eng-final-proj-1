```python
import os
import pandas as pd
```

```python
future50_df = pd.read_csv('/Users/ngan/PycharmProjects/data-eng-final-proj-1/clean_data/sales/Future50.csv')
independence100_df = pd.read_csv('/Users/ngan/PycharmProjects/data-eng-final-proj-1/clean_data/sales/Independence100.csv')
top250_df = pd.read_csv('/Users/ngan/PycharmProjects/data-eng-final-proj-1/clean_data/sales/top250.csv')
```

```python
future50_df = future50_df.loc[:,['Restaurant','Sales','City','State','YOY_Sales']]
future50_df['Sales']=future50_df['Sales']*100000
independence100_df = independence100_df.loc[:,['Restaurant','Sales','City','State']]
independence100_df['YOY_Sales']='0%'
top250_df = top250_df.loc[:,['Restaurant','Sales','City','State','YOY_Sales']]
top250_df['Sales']=top250_df['Sales']*100000
```

```python
display(future50_df.head(5))
display(independence100_df.head(5))
display(top250_df.head(5))
```

## Create state df

```python
states = ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut', 'Delaware', 'Florida', 'Georgia', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky', 'Louisiana', 'Maine', 'Maryland', 'Massachusetts', 'Michigan', 'Minnesota', 'Mississippi', 'Missouri', 'Montana', 'Nebraska', 'Nevada', 'New Hampshire', 'New Jersey', 'New Mexico', 'New York', 'North Carolina', 'North Dakota', 'Ohio', 'Oklahoma', 'Oregon', 'Pennsylvania', 'Rhode Island', 'South Carolina', 'South Dakota', 'Tennessee', 'Texas', 'Utah', 'Vermont', 'Virginia', 'Washington', 'West Virginia', 'Wisconsin', 'Wyoming']
state_df = pd.DataFrame({'state_id': range(1, len(states)+1), 'State': states})
state_df = state_df.sort_values(by='State')
display(state_df.head(5))
state_id_dict = dict(zip(state_df['State'], state_df['state_id']))
```

```python
future50_df['state_id'] = future50_df['State'].map(state_id_dict)
future50_df['state_id'] = future50_df['state_id'].fillna(0)
future50_df['state_id']=future50_df['state_id'].astype(int)

independence100_df['state_id'] = independence100_df['State'].map(state_id_dict)
independence100_df['state_id'] = independence100_df['state_id'].fillna(0)
independence100_df['state_id']=independence100_df['state_id'].astype(int)

top250_df['state_id'] = top250_df['State'].map(state_id_dict)
top250_df['state_id'] = top250_df['state_id'].fillna(0)
top250_df['state_id']=top250_df['state_id'].astype(int)

display(future50_df.head(5))
display(independence100_df.head(5))
display(top250_df.head(5))
```

```python
# Concat dataframes

business_df = pd.concat([future50_df,independence100_df,top250_df])
without_state_business_df = business_df.drop('State',axis=1)
display(without_state_business_df)
```
